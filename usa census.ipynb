# United States Census

The data here is for the "Census Income" dataset, which contains data on adults from the 1994 census. This data is labelled with whether the person's yearly income is above or below $50k.

### Problem statement 
Here we are looking at the main factors affecting the annual income of a USA citizen.


### Data description

The dataset consists of 32,562 rows and 14 features. The description of each feature is given below:


Discription about data:

-age: Tells about age of the person and continuous in nature.

-workclass: It defines whether the person belongs to Private, Self-emp-not-inc, Self-emp-inc, Federal-gov, Local-gov, State-gov, Without-pay, Never-worked in terms of way of income.

-fnlwgt: It shows financial status and it is continuous in nature.

-education:It shows education qulalification of person whether they are Bachelors, Some-college, 11th, HS-grad, Prof-school, Assoc-acdm, Assoc-voc, 9th, 7th-8th, 12th, Masters, 1st-4th, 10th, Doctorate, 5th-6th, Preschool.

-education-num:Continuous and represent the education qualification into numbers.

-marital-status: Married-civ-spouse, Divorced, Never-married, Separated, Widowed, Married-spouse-absent, Married-AF-spouse.

-occupation: Tech-support, Craft-repair, Other-service, Sales, Exec-managerial, Prof-specialty, Handlers-cleaners, Machine-op-inspct, Adm-clerical, Farming-fishing, Transport-moving, Priv-house-serv, Protective-serv, Armed-Forces.

-relationship: Wife, Own-child, Husband, Not-in-family, Other-relative, Unmarried.

-race: White, Asian-Pac-Islander, Amer-Indian-Eskimo, Other, Black.

-sex: Female, Male.

---Following Dataset attributes shows financial growth of person:

-capital-gain: continuous. Its the rise in the value of a capital asset (investment or real estate) that gives it a higher worth than the purchase price

-capital-loss: continuous. Its the loss incurred when a capital asset, such as an investment or real estate, decreases in value

-hours-per-week: continuous. Its the work hours per week

-native-country: United-States, Cambodia, England, Puerto-Rico, Canada, Germany, Outlying-US(Guam-USVI-etc), India, Japan, Greece, South, China, Cuba, Iran, Honduras, Philippines, Italy, Poland, Jamaica, Vietnam, Mexico, Portugal, Ireland, France, Dominican-Republic, Laos, Ecuador, Taiwan, Haiti, Columbia, Hungary, Guatemala, Nicaragua, Scotland, Thailand, Yugoslavia, El-Salvador, Trinadad&Tobago, Peru, Hong, Holand-Netherlands

### Data processing

We have to perform the cleaning of the data in order to analysis properly.

We saw that there are ? values in place of some data which are miscoded and doesn't tell anything so we have to replace them with null values and then with most frequent values also

```python
#replacing the incorrect data with the NaN values
df=df1.replace("?",np.nan)
df
```

```python
#finding the most frequent values
df.mode()
```

```python
#checking for the null values
df.isnull().sum()
```

```python
#handling null values
df["workclass"]=df["workclass"].fillna("Private")
df["occupation"]=df["occupation"].fillna("Prof-specialty")
df["native.country"]=df["native.country"].fillna("United-States")
df
```

```python
gender = {'Male': 1,'Female': 2}
df['sex'] = [gender[item] for item in df['sex']]
```

```python
''' we are converting income into integer values where above 50k is assinged as 1 and below 50k as 0'''
def income_to_numeric(x):
    if x=='>50K':
        return 1
    if x=='<=50K':
        return 0
    
df['income1'] = df['income'].apply(income_to_numeric)
```

## Bivariant analysis

So we are going to use above pairs for analysis

### age-income

```python
sns.boxplot(y='age',x='income1',hue='income',data=df)
sns.despine()
```

This shows that person among age group of 30 to 50 have more income than person among age group of 25 to 45
hence people above age 30 tend of have more income. 


### education.num-income
